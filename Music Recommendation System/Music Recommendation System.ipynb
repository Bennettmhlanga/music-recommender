# DEALING WITH ARTISTS NAME

```python
import re
import string

# Define a function to clean the artist names
def clean_artist_names(artist_list):
    cleaned_list = []
    for artists in artist_list:
        cleaned_artists = []
        for artist in artists:
            # Remove leading/trailing spaces and punctuation
            artist = artist.strip().translate(str.maketrans("", "", string.punctuation))
            cleaned_artists.append(artist)
        cleaned_list.append(" ".join(cleaned_artists))
    return cleaned_list

# Apply the function to the 'artists' column
df['artists'] = df['artists'].apply(lambda x: re.findall(r"'(.*?)'", x))  # Convert string representation to list
df['artists'] = clean_artist_names(df['artists'])
```

```python
# Define a function to clean the album names
def clean_album_names(album):
    # Remove punctuation from the album name
    album = album.translate(str.maketrans("", "", string.punctuation))
    # Replace multiple spaces with a single space
    album = re.sub(r'\s+', ' ', album)
    return album.strip()

# Clean the 'album' column
df['album'] = df['album'].apply(clean_album_names)
```
